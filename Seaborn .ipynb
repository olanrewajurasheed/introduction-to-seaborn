# Controlling plotted figures and Aesthetics

```python
def sinplot(flip =1):
    x=np.linspace(0,14,100)
    for i in range(1,7):
        plt.plot(x, np.sin(x+i*0.5)*(7-i)*flip)
```

```python
sinplot(-1)
```

```python
sns.set_style('whitegrid')
```

```python
sinplot()
```

```python
sns.set_style('white')
sinplot()
```

```python
sns.set_style('white')
sinplot()
sns.despine()
```

```python
sns.set_style('white')
sinplot()
sns.despine(left = True)
```

```python
sns.set_style('white')
sinplot()
sns.despine(left = True,  bottom = True)
```

```python
sns.set_style('ticks')
sinplot()
sns.despine(left = True,  bottom = True)
```

```python
sns.set_style('ticks', {'axes.grid': True, 'xtick.direction': 'in'})
sinplot()
sns.despine(left = True,  bottom = False, right = False)
```
